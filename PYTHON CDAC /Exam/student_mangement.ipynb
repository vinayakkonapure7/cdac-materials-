```python
class Student:
    def __init__(self, rollno, name, gpa, branch):
        self.rollno=rollno
        self.name=name
        self.gpa=gpa
        self.branch=branch
    def __str__(self):
        return f"Rollno: {self.rollno}, Name: {self.name}, Gpa: {self.gpa}, Branch: {self.branch}"
    def to_dict(self):
        return self.__dict__
```

```python
Student_list=[]
```

```python
def add_student():
    try:
        rollno=int(input("enter the rollno: "))
        for s in Student_list:
            if s.rollno == rollno:
                print("rollno already exists...")
                return
        name=input("enter the name: ")
        gpa=int(input("enter the gpa: "))
        branch=input("enter the branch: ")
        stud=Student(rollno,name,gpa,branch)
        Student_list.append(stud)
        print("student added sucessfully!!!")
    except:
        print("invalid input!!!!")
```

```python
add_student()
```

```python
def display_student():
    for s in Student_list:
        print(s)
display_student()
```

```python
def search_student():
    try:
        rollno=int(input("enter the rollno for search : "))
    except:
        print("Invalid data type for roll number. Please try again with numbers only.")
        return False
        
    rollnos=[s.rollno for s in Student_list]
    
    if rollno not in rollnos:
        print("This roll number does not exist in our database. Please try with another.")
        return False
        
    idx=rollnos.index(rollno)
    s=Student_list[idx]
    
    print("Student found...")
    print(f"Name: {s.name}\nGPA: {s.gpa}\nBranch: {s.branch}")
    return idx
```

```python
search_student()
```

```python
def delete_student():
    idx=search_student()
    if isinstance(idx,bool) and idx==False:
        return
    yn = input("Are you sure to delete this student? (yes/no): ")
    if yn.lower() == "yes":
        del Student_list[idx]
        print("Student record deleted successfully")
    else:
        print("delete canceled")
```

```python
delete_student()
```

```python
def update_student():
    idx=search_student()
    if isinstance(idx,bool) and idx==False:
        return
    s=Student_list[idx]
    print("enter the deatils for update...")
    name=input("enter the name: ")
    gpa=int(input("enter the gpa: "))
    branch=input("enter the branch: ")
    if name:
        s.name=name
    if gpa:
        s.gpa=gpa
    if branch:
        s.branch=branch
    print("upddated sucessfully!!!")
    display_student()
```

```python
update_student()
```

```python
display_student()
```

```python
def export_json():
    import json
    data=[s.to_dict() for s in Student_list]
    filename="students.json"
    with open(filename,"w") as file:
        json.dump(data,file,indent=3)
    print(f"exported json file sucessfully!!!{filename}")
```

```python
export_json()
```

```python
def export_csv():
    import csv
    data=[s.to_dict() for s in Student_list]
    filename="students.csv"
    with open(filename,"wt") as f1:
        csv_writer=csv.DictWriter(f1,fieldnames=["rollno", "gpa", "name", "branch"],lineterminator="\n")
        csv_writer.writeheader()
        csv_writer.writerows(data)
    print(f"exported csv file sucessfully!!!{filename}")
        
```

```python
def main():
    print("====== welcome to student mangement ======")
    while True:
        print("1.add student \n2.update student \n3.delete student \n4.search student \n5.view student \n6.export as [json or csv] \n7.exit")
        choice=int(input("enter the choice: "))
        if choice==1:
            add_student()
        elif choice==2:
            update_student()
        elif choice==3:
            delete_student()
        elif choice==4:
            search_student()
        elif choice==5:
            display_student()
        elif choice==6:
            ch=int(input("export as \n 1.json \n 2.csv \n enter the choice: "))
            if ch==1:
                export_json()
            elif ch==2:
                export_csv()
            else:
                print("invalid input")
        elif choice==7:
            print("loged out...")
            return
        else:
            print("invalid choice")
```

```python
if __name__=="__main__":
    main()
```
