```python
students=[]
```

```python
def add_student():
    try:
        rollno=int(input("enter the rollno: "))
        for s in students:
            if s["rollno"]==rollno:
                print("rollno exists...")
                return
        name=input("enter the name: ")
        gpa=float(input("enter the gpa: "))
        branch=input("enter the branch: ")
        student = {
                    "rollno": rollno,
                    "name": name,
                    "gpa": gpa,
                    "branch": branch
                }

        students.append(student)
        print("Student added successfully")
    except:
        print("invalid input")
```

```python
add_student()
```

```python
students
```

```python
def display_student():
    for s in students:
        print(s)
```

```python
display_student()
```

```python
def search_student():
    try:
        rollno=int(input("enter the rollno: "))
    except:
        print("must be intger")
        return False

    rollnos=[s["rollno"] for s in students]
    if rollno not in rollnos:
        print("Rollno does not exist")
        return False
    idx=rollnos.index(rollno)
    s=students[idx]

    print("student found...")
    print("name:",s["name"])
    print("gpa:",s["gpa"])
    print("branch:",s["branch"])
    return idx
 
```

```python
search_student()
```

```python
def delete_student():
    idx=search_student()
    if isinstance(idx,bool) and idx==False:
        return
    yn = input("Are you sure to delete this student? (yes/no): ")
    if yn.lower() == "yes":
        del students[idx]
    else:
        print("delete canceled")
```

```python
delete_student()
```

```python
display_student()
```

```python
def update_student():
    idx=search_student()
    s=students[idx]
    if isinstance(idx,bool) and idx==False:
        return
    print("enter the updated student details : ")
    name=input("enter the name: ")
    gpa=float(input("enter the gpa: "))
    branch=input("enter the branch: ")
    if name:
        s["name"]=name
    if gpa:
        s["gpa"]=gpa
    if branch:
        s["branch"]=branch
    print("updated sucessfully!!!")
    display_student()
```

```python
update_student()
```

```python
def export_json():
    import json
    filename="stud.json"
    with open(filename,"w") as file:
        json.dump(students,file,indent=3)
    print(f"Student data exported to JSON file - {filename}")
```

```python
export_json()
```

```python
def export_CSV():
    import csv
    filename="stud.csv"
    with open(filename,"wt") as f1:
        csv_writer=csv.DictWriter(f1,fieldnames=["rollno", "gpa", "name", "branch"],lineterminator="\n")
        csv_writer.writeheader()
        csv_writer.writerows(students)
    print(f"Student data exported to csv file - {filename}")
```

```python
export_CSV()
```

```python
def main():
    print("====== welcome to student mangement ======")
    while True:
        print("1.add student \n2.update student \n3.delete student \n4.search student \n5.view student \n6.export as [json or csv] \n7.exit")
        choice=int(input("enter the choice: "))
        if choice==1:
            add_student()
        elif choice==2:
            update_student()
        elif choice==3:
            delete_student()
        elif choice==4:
            search_student()
        elif choice==5:
            display_student()
        elif choice==6:
            ch=int(input("export as \n 1.json \n 2.csv \n enter the choice: "))
            if ch==1:
                export_json()
            elif ch==2:
                export_CSV()
            else:
                print("invalid input")
        elif choice==7:
            print("loged out...")
            return
        else:
            print("invalid choice")
```

```python
if __name__=="__main__":
    main()
```
