                            Word Frequency Counter

```python
sec = input("enter the sentence: ")

lower_sec = sec.lower()
words = lower_sec.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1  
    else:
        word_count[word] = 1

for key, value in word_count.items():
    print(f"{key} : {value}")
```

                            Dictionary Sorting Challenge

```python
stud = {
    "John": 85,
    "Alice": 92,
    "Bob": 78,
    "Emma": 95,
    "David": 88
}
stud
```

```python
asc=dict(sorted(stud.items(),key=lambda x:x[1]))
asc
```

```python
desc=dict(sorted(stud.items(),key=lambda x:x[1],reverse=True))
desc
```

```python
topper=max(stud,key=stud.get)
print("tooper is: ",topper,stud[topper])
```

```python
lowest=min(stud,key=stud.get)
print("lowest is: ",lowest,stud[lowest])
```

                            Banking Transaction Summary

```python
transactions = {
1: {"amount": 5000,"type": "credit"},
2: {"amount": 4000,"type": "debit"},
3: {"amount": 8000,"type": "credit"},
4: {"amount": 10000,"type": "debit"},
5: {"amount": 500,"type": "credit"},
}
transactions
```

```python
credit=[]
debit=[]
for id,details in transactions.items():
    val=details["type"] in 'credit'
    if val==True:
        cal=details['amount']
        credit.append(cal)
        print(f"{id}->{details}")
        sum=0
        for i in credit:
                sum+=i
    elif val == False:
        cal2=details['amount']
        debit.append(cal2)
        print(f"{id}->{details}")
        summ=0
        for j in debit:
            summ+=j 
print(f" total credit amount: {sum}")
print(f" total debit amount: {summ}")
print("number of credit transactions: ",len(credit))
print("number of debit transactions: ",len(debit))
```
