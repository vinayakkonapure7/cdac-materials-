```python
import sqlite3
```

```python
conn=sqlite3.connect("Pstudent.db")
cursor=conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS studentp(
id INTEGER PRIMARY KEY,
name TEXT,
age INTEGER ,
course TEXT
)
""")
```

```python
def add_stud(name,age,course):
    cursor.execute("INSERT INTO studentp(name,age,course) VALUES(?,?,?)",(name,age,course))
    conn.commit()
```

```python
def view_stud():
    cursor.execute("SELECT * FROM studentp")
    for row in cursor.fetchall():
        print(row)
```

```python
def update_stud(name,new_course):
    cursor.execute("UPDATE studentp SET course=? WHERE name=?",(new_course,name))
    conn.commit()
```

```python
def delete_stud(name):
    cursor.execute("DELETE FROM studentp WHERE name=?",(name,))
    print("deleted")
    conn.commit()
```

```python
add_stud("john",23,"AI")
```

```python
add_stud("ALICE",23,"AI")
```
