```python
import sqlite3
```

```python
conn=sqlite3.connect("Student.db")
cursor=conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students(
ID INTEGER PRIMARY KEY,
name TEXT,
age INTEGER,
course TEXT
)
""")
```

```python
def add_student():
    name=input("enter name of student: ")
    age=(input("enter the age: "))
    course=(input("enter the course: "))

    if not age.isnumeric():
        print("age must be integer")
        exit(1)
    age=int(age)
    if age >0 and age <=150:
        cursor.execute("INSERT INTO students (name,age,course) VALUES(?,?,?)",(name,age,course))
        print("added sucessfully...")
        conn.commit()
    else:
        print(f"{age} must under 0-150")
```

```python
def view_students():
    cursor.execute("SELECT * FROM students")
    for row in cursor.fetchall():
        print(row)
```

```python
def search_student():
    name=input("enter name of student : ")
    cursor.execute("SELECT * FROM students WHERE name=?",(name,))
    rows=cursor.fetchall()
    if rows:
        print(f"{name} details: ")
        # for row in rows:
        print(rows)
    else:
        print("not found..")
        
    
```

```python
def update_student():
    
    print("Update a student’s details \n 1.age  \n 2.course")
    ch=int(input("enter the choice: "))
    if ch==1:
        name=input("enter name of student which you want to update: ")
        age=int(input("enter the age: "))
        cursor.execute("UPDATE students SET age=? WHERE name=?",(age,name))
        print(f"updated {age} of student {name} ")
        conn.commit()
    elif ch==2:
        name=input("enter name of student which you want to update: ")
        course=(input("enter the course: "))
        cursor.execute("UPDATE students SET course=? WHERE name=?",(course,name))
        print(f"updated {course} of student {name} ")
        conn.commit()
    else:
        print("invalid choice...")
```

```python
def sort_student():
    print("sort  student’s details \n 1.age  \n 2.course")
    ch=int(input("enter the choice: "))
    if ch==1:
        cursor.execute("SELECT * FROM students ORDER BY age DESC")
        for row in cursor.fetchall():
            print(row)
    elif ch==2:
        cursor.execute("SELECT * FROM students ORDER BY course DESC")
        for row in cursor.fetchall():
            print(row)
    else:
        print("invalid choice...")
    
```

```python
def delete_student():
    name=input("enter name of students which you want to delete: ")
    cursor.execute("DELETE FROM students WHERE name=?",(name,))
    print("deleted...")
    conn.commit()
    
    
```

```python
def menu():
    print("----- Student Record Management System -----")
    while(True):
        print("====== menu =======")
        print("1. Add a new student record")
        print("2. Display all student records")
        print("3. Search for a student by name")
        print("4. sort a student’s details DSEC")
        print("5. Update a student’s details")
        print("6. Delete a student record")
        print("7. Exit the application")
        ch=int(input("enter the choice: "))
        if ch==1:
            add_student()
        elif ch==2:
            view_students()
        elif ch==3:
            search_student()
        elif ch==4:
            sort_student()
        elif ch==5:
            update_student()
        elif ch==6:
            delete_student()
        elif ch==7:
            print("loged out...")
            return
        else:
            print("invalid choice...")
    
```

```python
menu()
```
