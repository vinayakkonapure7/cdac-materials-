```python
class Product:
    def __init__(self,product_id,name,price,category,brand):
        self.product_id=product_id
        self.name=name
        self.price=price
        self.category=category
        self.brand=brand
    def __str__(self):
        return f"Product(ID={self.product_id},NAME={self.name},PRICE={self.price},CATEGORY={self.category},BRAND={self.brand})"

p1=Product(1,"apple",100,"fruit","kashmir")
print(p1)
```

```python
class product:
    def __init__(self,name,price):
        self.name=name
        self.__price=None
        self.set_price(price)
        
    def get_price(self):
        return self.__price
    def set_price(self,price):
        if price >0:
            self.__price=price
        else:
            print("price mustbe gretar than 0")
    def __str__(self):
        return f"product:(Name={self.name},Price={self.price})"
        
class ElectronicProduct(product):
    def __init__(self,name,price,warranty_years,power_usage):
        super().__init__(name,price)
        self.warranty_years=warranty_years
        self.power_usage=power_usage
    def __str__(self):
        return (f"ElectronicProduct={self.name}\n"
                f"Price={self.get_price()}\n"
                f"warranty_years={self.warranty_years}years\n"
                f"power_usage={self.power_usage}w\n")
    
        

        
```
