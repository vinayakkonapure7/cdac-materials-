Assignment 1: Student Record Management System Using Functions

```python
stud={}
```

```python
def add_stud():
    name=input("enter the Name of student: ")
    age=(input("enter the Age of student: "))
    if not age.isnumeric():
        print("age must be integer...")
        return
    age=int(age)
    
    marks=[] 
    for i in range(3):
        mul=(input(f"enter the Marks {i+1}: ")) 
        if not mul.isnumeric():
            print("marks must be integer...")
            return
        marks.append(int(mul))
    
    stud[name]={
            "age":age,
            "marks":marks
    }
    
    print("Student add sucessfully...")
```

```python
add_stud()
```

```python
stud.items()
```

```python
def update():
    up=input("enter the name of student for update : ")
    if up in stud:
        print("what you want to update \n 1.age \n 2.marks")
        ch=int(input("enter the choice[1 or 2]: "))
        if ch==1:
            age=(input("enter the Age of student: "))
            if not age.isnumeric():
                print("age must be integer...")
                return
            age=int(age)
            stud[up]['age']=age
            print("Age is updated....")
        elif(ch==2):
            marks=[]
            for i in range (3):
                mul=input(f"enter the marks{i+1}: ")
                if not mul.isnumeric():
                    print("marks must be integer...")
                    return
                marks.append(int(mul))
            stud[up]['marks']=marks
            print("Marks is updated...")
                
```

```python
def delete():
    name=input("enter the name of student for delete : ")
    if name in stud:
        del stud[name]
        print("student deleted succesfully...")
    else:
        print("not found...")
    
```

```python
def display():
    for name,details in stud.items():
        print(f"{name}-> {details}")
        
```

```python
def avg():
    enter=input("enter the name: ")
    for name,details in stud.items():
        if enter in name:
            avg=sum(details['marks'])/len(details['marks'])
            print(f"average of student {name} is {avg}")
```

```python
def menu():
    
    while(True):
        print("Enter your choice: ")
        print("1. Add a new student")
        print("2. Update an existing student’s marks or age")
        print("3. Delete a student record")
        print("4. Display all student records")
        print("5. display the average marks of a specific student")
        print("6. Exit")
        choice=int(input("enter the choice: "))
        if choice==1:
            add_stud()
        elif choice==2:
            update()
            # break
        elif choice==3:
            delete()
            # break
        elif choice==4:
            display()
        elif choice==5:
            avg()
            # break
        elif choice==6:
            print("exit..")
            return 
        else:
            print("enterd wrong choice...")
menu()
```

                        Assignment 2: Flexible Calculator Using Function Arguments

```python
def calculate(*args,opp):
    if opp=="add":
        return sum(args)
    elif opp =='sub':
        result=args[0]
        for i in args[1:]:
            result -= i
        return result
        
    elif opp=='div':
        result=args[0]
        for i in args[1:]:
            if i == 0:
                return("cant divide by 0")
            result /= i
        return result
    elif opp=='mul':
        result=args[0]
        for i in args[1:]:
            result *= i 
        return result
        
calculate(10,0,opp='div')

            
    
```

                            Data Persistence Using Pickling

```python
import pickle
```

```python
def save_data(filename,data):
    with open(filename,"wb") as file:
        pickle.dump(data,file)
```

```python
def load_data(filename):
    with open(filename,"rb") as file:
        return pickle.load(file)
```

```python
def menu():
    student={"name":"john","age":21}
    save_data("student.pkl",student)

    load=load_data("student.pkl")
    print(load)
```

```python
menu()
```
